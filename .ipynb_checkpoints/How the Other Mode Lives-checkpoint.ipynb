```python
# Define a few functions to keep the later code shorter and more readable
def plot_series(view, x_label, y_label, title, filename, y_as_percentage=False, savefig=True):
    """Plot a pandas Series using matplotlib, save it, and display it in the notebook.

    Args:
        view (pd.Series): Data series to plot.
        x_label (str): Label for the x-axis.
        y_label (str): Label for the y-axis.
        title (str): Title for the plot.
        filename (str): Name of the file to save the plot to.
        y_as_percentage (bool): Whether to format the y-axis values as percentages.
        savefig (bool): Whether to save the figure to a file.
    """
    plt.plot(view.index, view.values * 100 if y_as_percentage else view.values)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    if y_as_percentage:
        plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()]) 
    if savefig:
        plt.savefig(f'{filename}.jpg', bbox_inches='tight', dpi=300)
    plt.show()  # Display the plot in the notebook

def filter_df(df, age_range, column, condition_value):
    """
    Filters a DataFrame based on age and a specified condition in another column.
    
    Parameters:
    df (pandas.DataFrame): The DataFrame to be filtered.
    age_range (tuple): A tuple of two integers defining the range of ages to be included.
    column (str): The name of the column to be checked against the condition_value.
    condition_value (any): The value to be checked in the specified column. Rows with this value will be excluded.

    Returns:
    pandas.DataFrame: The filtered DataFrame.
    """
    filtered_df = df[(df['AGE'].between(age_range[0], age_range[1])) & (df[column] != condition_value)]
    return filtered_df

def plot_double_series(view1, view2, x_label, y_label, title, filename, yticks_percentage=False):
    """
    This function plots two data series on the same plot, sets the labels and title, formats the y-axis as percentages (if needed), adds a legend, saves the figure to a file, and also displays the figure in the Jupyter notebook.

    Args:
        view1 (pd.Series): First data series to plot.
        view2 (pd.Series): Second data series to plot.
        x_label (str): Label for the x-axis.
        y_label (str): Label for the y-axis.
        title (str): Title for the plot.
        filename (str): Name of the file to save the plot to.
        yticks_percentage (bool, optional): Whether to format the y-axis values as percentages. Defaults to False.
    """
    plt.plot(view1.index, view1.values * 100, label="Women")
    plt.plot(view2.index, view2.values * 100, label="Men")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    if yticks_percentage:
        plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()])
    plt.savefig(f'{filename}.jpg', bbox_inches='tight', dpi=300)
    plt.show()  # Display the plot in the notebook

def filter_and_group_df(df, column, condition_value, age_range, groupby_column, aggregate_column, aggregate_function):
    """
    Filters, groups, and aggregates data from a DataFrame based on specified parameters.

    Parameters:
    df (pandas.DataFrame): The DataFrame to be filtered, grouped, and aggregated.
    column (str): The name of the column to which the condition will be applied.
    condition_value (any): The value to be compared with the aggregate_column.
    age_range (tuple): A tuple of two integers specifying the range of the index to be included in the output.
    groupby_column (str): The name of the column by which the DataFrame will be grouped.
    aggregate_column (str): The name of the column that will be aggregated.
    aggregate_function (str): The name of the function that will be applied to aggregate the data.

    Returns:
    pandas.DataFrame: The filtered, grouped, and aggregated DataFrame.
    """
    df.loc[:, column] = df[aggregate_column] > condition_value
    view = df.groupby(groupby_column)[column].apply(aggregate_function)
    view = view[(view.index > age_range[0]) & (view.index <= age_range[1])]
    return view
```

## Understand and prep dataset

Show the variable names and (lack of) missingness. Missing values come pre-imputed by Census. 

In this blog, we'll accept the Census-provided imputation and ignore sample weights for brevity.

Prepare subsets of the raw data to be used for modeling/plotting

```python
# Split data into men and women
df_women = raw_data[raw_data['SEX'] == 2].copy()  # male is 1, female is 2
df_men = raw_data[raw_data['SEX'] == 1].copy()  # male is 1, female is 2

# Indicator for women who have given birth, between 15-50 years old
df_women.loc[:, 'ISAMOM'] = df_women['FRAGE1'] != 999 

# Subset of women aged 36-50 - see method note below for how this deals with a missing data problem
df_women_36_50 = filter_df(df_women, (36, 50), 'FRAGE1', 999)

# Dataframe of only moms, with variable for age of eldest biological child
df_moms = raw_data[raw_data['FRAGE1'] != 999]  # FRAGE1: age of women in years at birth of first child, only asked if woman is 15-50 at time of survey
df_moms['KIDAGE'] = df_moms['YEAR'] - df_moms['FRBIRTHY1']
```

Check survey size

```python
person_count = len(raw_data)
mom_survey_frame_count = len(df_women[df_women['AGE'].between(15,50)])
household_count = len(raw_data['SERIAL'].unique())
mom_count = df_women['ISAMOM'].value_counts()[1]
print(f"Number of surveyed people: {person_count}")
print(f"Number of surveyed households: {household_count}")
print(f"Number of surveyed women aged 15-50: {mom_survey_frame_count}")
print(f"Number of surveyed biological mothers aged 15-50: {mom_count}")
```

#### Method note: 
Notice the **missing data** problem here. Because we cannot observe future events, we don't know the "age of first birth" for women who have not (yet) given birth.  That's why we only show age at first birth for women who were aged 36–50 when surveyed and stop the plot at age 35. Younger women’s responses are left out to avoid skewing the curve deceptively to the left by omitting future births. 

```python
view2 = df_women_36_50.groupby('FRAGE1')['FRAGE1'].count()
view2 = view2 / sum(view2)

# Age is reported in months
view2 = view2[view2.index < 36 * 12]

# Convert Index to Series for plotting.
view2_for_plot = pd.Series(data=view2.values, index=view2.index / 12)

plot_series(view2_for_plot, 'Age at first birth, for respondents >35 y/o', '', 
            'Age at first birth', 'age_at_first_birth')
```

## Question #2: How is family income during childhood influenced by a mother's age?

```python
df_moms['kidage_bucket'] = pd.cut(
    df_moms['KIDAGE'], bins=[-1, 2, 5, 8, 11, 14, 17, float('inf')],
    labels=['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18+']
)
df_moms['birthage_bucket'] = pd.cut(
    df_moms['FRAGE1'], bins=[216, 264, 312, 360, float('inf')],
    labels=['19-22', '23-26', '27-30', '30+']
)

income_mapping = {
    100: 2500,  # LESS THAN $5,000, using 2500 as a representative value
    210: 6250,  # 5,000 TO 7,499
    300: 8750,  # 7,500 TO 9,999
    430: 11250,  # 10,000 TO 12,499
    470: 13750,  # 12,500 TO 14,999
    500: 17500,  # 15,000 TO 19,999
    600: 22500,  # 20,000 TO 24,999 
    710: 27500,  # 25,000 TO 29,999
    720: 32500,  # 30,000 TO 34,999
    730: 37500,  # 35,000 TO 39,999
    740: 45000,  # 40,000 TO 49,999
    820: 55000,  # 50,000 TO 59,999
    830: 67500,  # 60,000 TO 74,999
    841: 87500,  # 75,000 TO 99,999
    842: 125000,  # 100,000 TO 149,999
    843: 150000,  # 150,000 OR MORE, using 150000 as a representative value
}

df_moms['family_income'] = df_moms['FAMINC'].map(income_mapping)

# Create a dataframe of average family income by mom's age at first birth and eldest child's age
income_buckets = df_moms.groupby(['KIDAGE', 'birthage_bucket'])['family_income'].agg(['mean', 'count'])

# Drop cells with less than 100 observations because they're too erratic
income_buckets = income_buckets[income_buckets['count'] > 100]

# Reset the index of the DataFrame so 'KIDAGE' and 'birthage_bucket' become regular columns
income_buckets = income_buckets.reset_index()
income_buckets = income_buckets.sort_values(['birthage_bucket', 'KIDAGE'], ascending=True)

# Create a figure and axis
fig, ax = plt.subplots()

# Loop over each unique value in 'birthage_bucket'
for bucket in income_buckets['birthage_bucket'].unique():
    # Subset the data to this bucket
    subset = income_buckets[income_buckets['birthage_bucket'] == bucket]
    # Plot the mean income against 'KIDAGE' for this bucket
    sns.lineplot(x='KIDAGE', y='mean', data=subset, label=bucket, ax=ax)

# Add a legend, labels, and a title
ax.legend(title="Mom's age at first birth")
ax.set_xlabel('Age of eldest child')
ax.set_ylabel('Mean family income')
ax.set_title('Family Income Trends During Childhood \nby Age of Mother')

# Format y-axis tick labels with commas
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))

# Show the plot
plt.savefig('income_dynamics.jpg', bbox_inches='tight', dpi=300)
```

### Method note:

Median is a more typical way to summarize income groups since the distribution is right-skewed. However, because the survey only permits income range responses, plotting the median makes the trends less smooth and harder to read. That's why I chose mean. The overall conclusions of median are similar, but the levels are lower.

## Question #3: How likely are moms to live with their firstborn child at home? Does this depend on the age gap between mother and child?

```python
# Firstborn child in home?
# Only include moms whose eldest is still a minor
# Dataframe of moms with a minor child 
df_moms_of_minors = df_moms[df_moms['KIDAGE'] < 18][['kidage_bucket', 'birthage_bucket', 'KIDAGE', 'ELDCH']]

# KIDAGE must be greater than ELDCH by more than one
# Necessary because of mismatch in how variables are observed, by calendar year vs current age
df_moms_of_minors['kid_not_home'] = df_moms_of_minors['KIDAGE'] - df_moms_of_minors['ELDCH'] > 1

inhome_buckets = df_moms_of_minors.groupby(['kidage_bucket', 'birthage_bucket'])['kid_not_home'].agg(['mean', 'count'])
```

```python
# Create plot
crosstab = pd.crosstab(df_moms_of_minors['birthage_bucket'], df_moms_of_minors['kid_not_home'], normalize='index')

# Select the True column values
true_values = crosstab[True] *100

# Plotting the bar chart
true_values.plot(kind='bar')

# Set the labels and title
plt.xlabel("Mother's age at birth of first child")
#plt.ylabel('Percent of Eldest Children not in Mother's Household')
plt.title("Percent of Eldest Children \nMissing from Mother's Household")
plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()])

# Display the plot
#plt.show()
plt.savefig('missing_eldest.jpg', bbox_inches='tight', dpi=300)
```

```python
# Create an empty dictionary to store the crosstab results
crosstab_dict = {} 

# Do crosstabs by child age range
for group in df_moms['kidage_bucket'].unique():
    print(group)
    buckets = df_moms_of_minors[df_moms_of_minors['kidage_bucket'] == group]
    crosstab_dict[group] = pd.crosstab(buckets['birthage_bucket'], buckets['kid_not_home'], normalize='index')
    print(crosstab_dict[group])

# Set up the figure and axes
fig, axes = plt.subplots(nrows=len(crosstab_dict) // 2, ncols=2, figsize=(10, 10), sharey=True)

# Plot the bar charts in the subplots and add subtitles
axes[0, 0].set_title('Children Aged 0-2')
axes[0, 1].set_title('Children Aged 3-5')
axes[1, 0].set_title('Children Aged 6-8')
axes[1, 1].set_title('Children Aged 9-11')
axes[2, 0].set_title('Children Aged 12-14')
axes[2, 1].set_title('Children Aged 15-17')

crosstab_dict['0-2'][True].plot(kind='bar', ax=axes[0, 0])
crosstab_dict['3-5'][True].plot(kind='bar', ax=axes[0, 1])
crosstab_dict['6-8'][True].plot(kind='bar', ax=axes[1, 0])
crosstab_dict['9-11'][True].plot(kind='bar', ax=axes[1, 1])
crosstab_dict['12-14'][True].plot(kind='bar', ax=axes[2, 0])
crosstab_dict['15-17'][True].plot(kind='bar', ax=axes[2, 1])

# Remove x-axis labels
axes[0, 0].set_xlabel('')
axes[0, 1].set_xlabel('')
axes[1, 0].set_xlabel('')
axes[1, 1].set_xlabel('')
axes[2, 0].set_xlabel('')
axes[2, 1].set_xlabel('')

# Set the common title
fig.suptitle("Missing Eldest Children, \nby Child Age and Parent's Age at Birth")

# Multiply y-axis tick labels by 100 and add percentage sign
for ax in axes.flat:
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}%'.format(x * 100)))

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.15, hspace=0.5)

# Display the plot
plt.savefig('missing_eldest_age_groups.jpg', bbox_inches='tight', dpi=300)
```
